## Load curated data

```python
input_file_path = 'data/data_curated.csv'
data_curated = pd.read_csv(input_file_path)
```

```python
data_curated.dropna().YearsExperience.value_counts().plot.pie(figsize=(5, 3))
```

```python
data_curated.dropna().CompanySize.value_counts().plot.pie(figsize=(5, 3))
```

```python
data_curated.dropna().WorkLocation.value_counts().plot.pie(figsize=(5, 3))
```

```python
data_curated.dropna().SuperivisionRole.value_counts().plot.pie(figsize=(5, 3))
```

```python
data_curated.dropna().WorkOutsideMainJob.value_counts().plot.pie(figsize=(5, 3))
```

```python
data_curated.dropna().Sex.value_counts().plot.pie(figsize=(5, 3))
```

```python
data_curated.dropna().CityLive.value_counts().plot.bar(figsize=(15, 5))
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data_curated.corr(), annot=False, linewidths=.005)
plt.show()
```
